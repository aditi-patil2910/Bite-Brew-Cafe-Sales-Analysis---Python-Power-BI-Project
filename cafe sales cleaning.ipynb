## Data Cleaning

```python
# standardizing column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df.columns.tolist()
```

```python
# to identify the wrong values
df['item'].unique()
```

```python
#-- Items ---
df['item']= df['item'].str.strip()
df['item'] = df['item'].replace({'ERROR': pd.NA, 'UNKNOWN': pd.NA,'': pd.NA})
df['item'] = df['item'].fillna('Others')
df['item'].head(10)
df['item'].unique()
```

```python
#-- Quantity ---
df['quantity'] = df['quantity'].replace({'ERROR':pd.NA, 'UNKNOWN':pd.NA, '':pd.NA})
df['quantity'] =  pd.to_numeric(df['quantity'], errors = 'coerce')
df['quantity'] = df['quantity'].fillna(df['quantity'].median())
df['quantity'].head(10)
df['quantity'].isnull().sum()
```

```python
#-- Price per unit
df['price_per_unit'] = df['price_per_unit'].replace({'ERROR':pd.NA, 'UNKNOWN':pd.NA, '':pd.NA})
df['price_per_unit'] = pd.to_numeric(df['price_per_unit'], errors='coerce')
df['price_per_unit'] = df['quantity'].fillna(df['quantity'].median())
df['price_per_unit'].head()
```

```python
# total spent
df['total_spent'] = df['total_spent'].replace({'ERROR':pd.NA, 'UNKNOWN':pd.NA, '':pd.NA})
df['total_spent'] = pd.to_numeric(df['total_spent'], errors='coerce')
df['total_spent'] = (df['quantity'] * df['price_per_unit']).round(2)
df['total_spent'].head()
```

```python
df['payment_method'].unique()
```

```python
#payment method 
df['payment_method'] = df['payment_method'].replace({'ERROR':pd.NA, 'UNKNOWN':pd.NA, 'Error':pd.NA, 'Nan':pd.NA, '':pd.NA})
df['payment_method'] = df['payment_method'].fillna('Unknown')
most_common = df[df['payment_method'] != 'Unknown']['payment_method'].mode()[0]
df['payment_method'] = df['payment_method'].replace('Unknown', most_common)
df['payment_method'] = df['payment_method'].str.strip().str.title()
df['payment_method'].head()
df['payment_method'].isnull().sum()
```

```python
#location
df['location'] = df['location'].replace({'ERROR':pd.NA, 'UNKNOWN':pd.NA, '':pd.NA})
df['location'] = df['location'].fillna('Unknown')
mode_location = df[df['location'] != 'Unknown']['location'].mode()[0]
df['location'] = df['location'].replace('Unknown', mode_location)
df['location'] = df['location'].str.strip().str.title()
df['location'].head()
df['location'].unique()
```

```python
# transaction_date
df['transaction_date'] = df['transaction_date'].replace({'ERROR':pd.NA, 'UNKNOWN':pd.NA, '':pd.NA})
df['transaction_date'] = pd.to_datetime(df['transaction_date'], errors = 'coerce')
df = df.sort_values('transaction_date')
df['transaction_date'] = df['transaction_date'].ffill()
```

## Saving the clean csv file 


```python
df.to_excel('Cafe_Sales_Clean.xlsx', sheet_name = 'Clean_Data', index = False)
```
